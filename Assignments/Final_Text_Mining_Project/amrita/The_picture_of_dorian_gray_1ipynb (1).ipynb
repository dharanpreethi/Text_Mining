```python
#WORDCLOUD

# Install necessary libraries
!pip install wordcloud

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from google.colab import files

# Upload a text file
uploaded = files.upload()

# Get the filename from the uploaded files
filename = list(uploaded.keys())[0]

# Read the text file
with open(filename, 'r', encoding='utf-8', errors='ignore') as f:
    text = f.read()

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400,
                      background_color='white',
                      stopwords=STOPWORDS,
                      min_font_size=10).generate(text)

# Display the generated image:
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# A pie chart of words used in an uploade text file

from collections import Counter
import matplotlib.pyplot as plt

# Assuming 'text' variable from the previous code block contains the file content

# Tokenize the text (split into words)
words = text.lower().split()

# Remove stop words (optional)
stop_words = set(STOPWORDS) # Use the STOPWORDS set from wordcloud
words = [word for word in words if word not in stop_words and word.isalnum()]

# Count word frequencies
word_counts = Counter(words)

# Get the top N most frequent words (adjust N as needed)
top_n = 10
most_common_words = word_counts.most_common(top_n)

# Extract labels and values for the pie chart
labels, values = zip(*most_common_words)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title(f"Top {top_n} Most Frequent Words")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#  a bar graph of 100 most used words

!pip install nltk

import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
# Download the 'averaged_perceptron_tagger_eng' resource
nltk.download('averaged_perceptron_tagger_eng')
import matplotlib.pyplot as plt
from collections import Counter
from google.colab import files
from nltk.corpus import stopwords
nltk.download('stopwords')


uploaded = files.upload()
filename = list(uploaded.keys())[0]

with open(filename, 'r', encoding='utf-8', errors='ignore') as f:
    text = f.read()

words = nltk.word_tokenize(text.lower())
stop_words = set(stopwords.words('english'))
words = [word for word in words if word not in stop_words and word.isalnum()]


tagged_words = nltk.pos_tag(words)
adjectives_and_nouns = [word for word, tag in tagged_words if tag.startswith('JJ') or tag.startswith('NN')]

word_counts = Counter(adjectives_and_nouns)

most_common_words = word_counts.most_common(100)
labels, values = zip(*most_common_words)

plt.figure(figsize=(20, 8)) # Adjust figure size as needed
plt.bar(labels, values)
plt.xticks(rotation=90)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("100 Most Frequent Adjectives and Nouns")
plt.show()
```
