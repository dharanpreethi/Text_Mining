#This code is for saving all NER of MY text file  into CSV too better analysis.

```python
import spacy
import csv

# Load the spaCy transformer model
nlp = spacy.load("en_core_web_trf")

# Read the Ramayana text
with open("ramayana.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Process the text
doc = nlp(text)

# Open CSV file to save results
with open("ramayana_named_entities.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Entity Text", "Entity Label"])  # Header

    # Extract and write named entities
    for ent in doc.ents:
        writer.writerow([ent.text, ent.label_])

print("Named entities saved to 'ramayana_named_entities.csv'.")
```

#This code is  for saving Persons Name through NER in CSV file

```python
import spacy
import csv

# Load the spaCy transformer model
nlp = spacy.load("en_core_web_trf")

# Read the Ramayana text
with open("ramayana.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Process the text
doc = nlp(text)

# Open CSV file to save only person names
with open("ramayana_person_names.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Person Name"])  # Header

    # Extract and write only PERSON entities
    for ent in doc.ents:
        if ent.label_ == "PERSON":
            writer.writerow([ent.text])

print("Person names saved to 'ramayana_person_names.csv'.")
```

#This code if for top 10 Frequent Names in Ramayana.
* Throgh this we can see in bar graph in top 10 persons discussed in Rmayana is mostly Male. Only 2 women is here.

*This shows the male dominant men hold more power, authority, or influence than women.






```python
import csv
from collections import Counter
import matplotlib.pyplot as plt

# Count frequency of person names
person_name_counts = Counter()

# Read the CSV and count names
with open("ramayana_person_names.csv", "r", encoding="utf-8") as infile:
    reader = csv.reader(infile)
    next(reader)  # Skip header
    for row in reader:
        name = row[0].strip()
        person_name_counts[name] += 1

# Get top 10 most frequent person names
top_names = person_name_counts.most_common(10)
names, counts = zip(*top_names)

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(names, counts, color='coral')
plt.xlabel("Person Names")
plt.ylabel("Frequency")
plt.title("Top 10 Person Names in Ramayana")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#This code is for Extracting the uniques Names form the Persons name
# I am extracting unique names because in the persons name i got repeted names.

```python
import csv

# To store unique names with gender
unique_persons = {}

# Read the existing file
with open("ramayana_persons_gender.csv", "r", encoding="utf-8") as infile:
    reader = csv.DictReader(infile)
    for row in reader:
        name = row["Person Name"].strip()
        gender = row["Gender"].strip()
        # Only store the first occurrence of each name
        if name not in unique_persons:
            unique_persons[name] = gender

# Write the unique names to a new CSV
with open("unique_ramayana_persons_gender.csv", "w", newline="", encoding="utf-8") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(["Person Name", "Gender"])
    for name, gender in unique_persons.items():
        writer.writerow([name, gender])

print("Unique person names saved to 'unique_ramayana_persons_gender.csv'.")
```

# This code for visualization of Male and Female Through Pie chart.

```python
import csv
from collections import Counter
import matplotlib.pyplot as plt

# Replace with your actual filename (with correct quotes)
csv_filename = 'unique_ramayana_persons_gender (4).csv'

# Count genders
gender_counts = Counter()

with open(csv_filename, 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        gender = row['Gender'].strip()
        if gender:
            gender_counts[gender] += 1

# Prepare data
labels = list(gender_counts.keys())
sizes = list(gender_counts.values())
colors = ['lightblue', 'lightpink']

# Plot pie chart
plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=180, colors=colors)
plt.title('Gender Distribution of Person Names in Ramayana')
plt.axis('equal')  # Equal aspect ratio ensures a perfect circle
plt.show()
```
