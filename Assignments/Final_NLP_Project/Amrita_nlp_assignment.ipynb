```python
#  NAMED ENTITY RECOGNITION OF AN UPLOADED FILE

!pip install spacy
!python -m spacy download en_core_web_sm

import spacy

# Load the pre-trained spaCy model
nlp = spacy.load("en_core_web_sm")

def ner_from_file(filepath):
    try:
        with open(filepath, "r", encoding="utf-8") as file:  # Handle potential encoding issues
            text = file.read()
    except FileNotFoundError:
        return "File not found."
    except Exception as e:  # Catch other potential errors
        return f"An error occurred: {e}"

    doc = nlp(text)

    entities = []
    for ent in doc.ents:
        entities.append((ent.text, ent.label_))

    return entities

# Example usage (replace 'your_file.txt' with your file path):
filepath = "/content/the-picture-of-dorian-gray.txt"
ner_results = ner_from_file(filepath)

if isinstance(ner_results, list): #Check if the result is a list of entities
  for entity, label in ner_results:
    print(f"Entity: {entity}, Label: {label}")
else:
  print(ner_results) # Print error message if file not found
```

```python
#TEXT SUMMARIZER OF A TEXT FILE USING SPACY

import spacy

def summarize_text(text, n_sentences=3):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)
    sentences = list(doc.sents)
    sentences_ranked = sorted(sentences, key=lambda s: len(s), reverse=True) #Rank by sentence length
    summary = " ".join([sent.text for sent in sentences_ranked[:n_sentences]])
    return summary

def summarize_from_file(filepath, n_sentences=3):
    try:
        with open(filepath, "r", encoding="utf-8") as file:
            text = file.read()
        summary = summarize_text(text, n_sentences)
        return summary
    except FileNotFoundError:
        return "File not found."
    except Exception as e:
        return f"An error occurred: {e}"

# Example usage
filepath = "/content/the-picture-of-dorian-gray.txt"
summary = summarize_from_file(filepath)
summary
```

```python
# PART OF SPEECH TAGGING USING ROBERTA AND SPACY USING SPACY

import spacy

def pos_tagging_from_file(filepath):
    try:
        with open(filepath, "r", encoding="utf-8") as file:
            text = file.read()
    except FileNotFoundError:
        return "File not found."
    except Exception as e:
        return f"An error occurred: {e}"

    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)

    pos_tags = []
    for token in doc:
        pos_tags.append((token.text, token.pos_))
    return pos_tags

# Example usage
filepath = "/content/the-picture-of-dorian-gray.txt"
pos_tags = pos_tagging_from_file(filepath)

if isinstance(pos_tags, list):
    for token, pos_tag in pos_tags:
        print(f"Token: {token}, POS Tag: {pos_tag}")
else:
    print(pos_tags) # Added indentation and print statement to handle the else case
```

```python
#  COLOURFUL PART OF SPEECH TAGGING

def colorful_pos_tagging_from_file(filepath):
    try:
        with open(filepath, "r", encoding="utf-8") as file:
            text = file.read()
    except FileNotFoundError:
        return "File not found."
    except Exception as e:
        return f"An error occurred: {e}"

    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)

    pos_tags = []
    for token in doc:
        color_map = {
            "NOUN": "\033[94m",  # Blue
            "VERB": "\033[92m",  # Green
            "ADJ": "\033[93m",   # Yellow
            "ADV": "\033[91m",   # Red
            "PRON": "\033[95m",  # Magenta
            "ADP": "\033[96m",   # Cyan
            "DET": "\033[90m",    # Gray
        }
        color_code = color_map.get(token.pos_, "\033[0m") # Default to no color if POS tag is not in the map
        pos_tags.append((token.text, token.pos_, color_code))

    return pos_tags

# Example Usage
filepath = "/content/the-picture-of-dorian-gray.txt"
colorful_pos_tags = colorful_pos_tagging_from_file(filepath)

if isinstance(colorful_pos_tags, list):
    for token, pos_tag, color_code in colorful_pos_tags:
        print(f"Token: {color_code}{token}\033[0m, POS Tag: {pos_tag}")  # Reset color after printing
else:
    print(colorful_pos_tags) # Added indentation and print statement to handle the else case
```

```python
#  dependency graphs in nlp using bert]

!pip install networkx matplotlib

import networkx as nx
import matplotlib.pyplot as plt

def dependency_graph(text):
    nlp = spacy.load("en_core_web_sm")
    doc = nlp(text)

    graph = nx.DiGraph()
    for token in doc:
        graph.add_node(token.text)
        if token.head.text != token.text:  #Avoid self-loops
            graph.add_edge(token.head.text, token.text, label=token.dep_)
    return graph


def plot_dependency_graph(graph):
    pos = nx.spring_layout(graph)  # Positions for nodes
    nx.draw(graph, pos, with_labels=True, node_size=1000, node_color="skyblue", font_size=10, font_weight="bold")
    edge_labels = nx.get_edge_attributes(graph, "label")
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)
    plt.title("Dependency Graph")
    plt.show()

def analyze_from_file(filepath, num_sentences=3):
  try:
      with open(filepath, "r", encoding="utf-8") as file:
          text = file.read()
      # Process only the first few sentences to make the graph readable
      sentences = list(nlp(text).sents)
      processed_text = " ".join([sent.text for sent in sentences[:num_sentences]])
      graph = dependency_graph(processed_text)
      plot_dependency_graph(graph)
  except FileNotFoundError:
      print("File not found.")
  except Exception as e:
      print(f"An error occurred: {e}")


# Example usage
filepath = "/content/the-picture-of-dorian-gray.txt"
analyze_from_file(filepath)
```

```python
#  GRAPH TO SHOW SEMANTIC RELATIONSHIP BETWEEN WORDS IN THE TEXT FILE USING BERT

!pip install transformers

from transformers import pipeline
import networkx as nx
import matplotlib.pyplot as plt

def semantic_similarity_graph(text, top_n=5):
    # Use a pre-trained model for semantic similarity
    nlp = pipeline("feature-extraction", model="bert-base-uncased")

    # Get word embeddings (features) from BERT
    features = nlp(text)

    # Create a graph to represent semantic relationships
    graph = nx.Graph()

    # Add nodes (words) to the graph
    words = text.split()
    for i in range(min(len(words), len(features[0]))):
      graph.add_node(i)

    # Calculate and add weighted edges based on cosine similarity
    from sklearn.metrics.pairwise import cosine_similarity

    for i in range(min(len(words), len(features[0]))):
      for j in range(i+1, min(len(words), len(features[0]))):
          similarity = cosine_similarity([features[0][i]], [features[0][j]])[0][0]
          if similarity > 0.5: # Only add edges if similarity is above a threshold
              graph.add_edge(i,j, weight=similarity)

    return graph

def plot_semantic_graph(graph, text):
    pos = nx.spring_layout(graph)
    edges = graph.edges()
    weights = [graph[u][v]['weight'] for u,v in edges]

    # Get the words from the text
    words = text.split()

    # Node labels
    labels = {i: words[i] for i in range(min(len(words), len(graph.nodes)))}

    nx.draw(graph, pos, with_labels=True, labels=labels, node_size=50, node_color="skyblue", font_size=8, font_weight="bold", width=weights)
    plt.title("Semantic Similarity Graph")
    plt.show()

def analyze_from_file(filepath, num_sentences=3):
    try:
        with open(filepath, "r", encoding="utf-8") as file:
            text = file.read()
        sentences = list(nlp(text).sents) # Assuming nlp is defined elsewhere (e.g., from spacy)
        processed_text = " ".join([sent.text for sent in sentences[:num_sentences]])
        graph = semantic_similarity_graph(processed_text)
        plot_semantic_graph(graph, processed_text) # Pass the text to the plotting function

    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

filepath = "/content/the-picture-of-dorian-gray.txt"
analyze_from_file(filepath)
```

```python
# semantic  relationship graph using bert

import spacy
import networkx as nx
import matplotlib.pyplot as plt
from transformers import pipeline
from sklearn.metrics.pairwise import cosine_similarity

# Load a larger spaCy model for better accuracy
!python -m spacy download en_core_web_lg
nlp = spacy.load("en_core_web_lg")

def semantic_similarity_graph(text, top_n=10, threshold=0.6): # Increased top_n and added a threshold
    # Use a larger BERT model for improved embeddings
    nlp_bert = pipeline("feature-extraction", model="bert-large-uncased")

    features = nlp_bert(text)
    graph = nx.Graph()

    words = [token.text for token in nlp(text) if not token.is_stop and not token.is_punct] #Filter out stop words and punctuation
    for i in range(min(len(words), len(features[0]))):
        graph.add_node(words[i])  # Use words as node labels directly


    for i in range(min(len(words), len(features[0]))):
        for j in range(i + 1, min(len(words), len(features[0]))):
            similarity = cosine_similarity([features[0][i]], [features[0][j]])[0][0]
            if similarity >= threshold: # Added threshold
                graph.add_edge(words[i], words[j], weight=similarity)

    return graph

def plot_semantic_graph(graph):
    plt.figure(figsize=(12, 12)) # Increased figure size for better visualization
    pos = nx.spring_layout(graph, k=0.5) # Adjust k for node spacing
    edges = graph.edges()
    weights = [graph[u][v]['weight'] for u,v in edges]

    nx.draw(graph, pos, with_labels=True, node_size=1500, node_color="maroon", font_size=10, font_weight="bold", width=weights, edge_color="gray") # Increased node size and font size

    # Add edge labels for better clarity (optional)
    edge_labels = nx.get_edge_attributes(graph,'weight')
    nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels, font_size=8)

    plt.title("Semantic Similarity Graph (BERT)", fontsize=14)
    plt.show()


def analyze_from_file(filepath, num_sentences=5): # Increased sentences for more context
    try:
        with open(filepath, "r", encoding="utf-8") as file:
            text = file.read()
        sentences = list(nlp(text).sents)
        processed_text = " ".join([sent.text for sent in sentences[:num_sentences]])
        graph = semantic_similarity_graph(processed_text)
        plot_semantic_graph(graph)

    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

filepath = "/content/the-picture-of-dorian-gray.txt"
analyze_from_file(filepath)
```
